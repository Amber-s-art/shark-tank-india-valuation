```python
import numpy as np 
import pandas as pd
```

```python
data = pd.read_csv('data\Shark Tank India.csv')
display(data.head())
display(data.columns)
```

```python
column_to_work_on = [
    # --- Context & Identifiers ---
    "Season Number", "Season Start", "Industry", "Started in", "Pitchers Average Age", 
    
    # --- Financial Features (Inputs) ---
    "Yearly Revenue", "Monthly Sales", "Gross Margin", "Net Margin", "EBITDA", 
    "Has Patents", "Bootstrapped", "SKUs",
    
    # --- Founder Expectations (The Anchor) ---
    "Original Ask Amount", "Original Offered Equity", "Valuation Requested",
    
    # --- Negotiation Results (Avoid using these to PREDICT Valuation) ---
    "Received Offer", "Accepted Offer", "Number of Sharks in Deal", 
    "Total Deal Amount", "Total Deal Equity","Total Deal Debt",
    "Debt Interest", "Deal Valuation"
]
df = data[column_to_work_on]

display(df.head())
```

```python
display(df.isnull().sum())
display(df.shape)
```

```python
# Convert 'Season Start' to a datetime object so we can extract the year
# we use 'coerce' to handle any invalid date formats gracefully, which will convert them to NaT (Not a Time)
df['Season Start'] = pd.to_datetime(df['Season Start'], errors='coerce')
```

```python
# Clean the Industry column so grouping works perfectly
df['Industry'] = df['Industry'].fillna('Others').str.strip().str.title()
```

```python
# Calculate the median starting year for each specific industry
industry_start_medians = df.groupby('Industry')['Started in'].transform('median')

# Fill the missing 'Started in' values with their respective Industry Median
df['Started in'] = np.where(
    df['Started in'].isna(),
    industry_start_medians,
    df['Started in']
)

# Crucial Step: Convert the years from decimals (e.g., 2019.0) to solid integers (2019)
df['Started in'] = df['Started in'].astype(int)
```

```python
# 2. Extract Season Year from 'Season Start' for age calculation
df['Season_Year'] = df['Season Start'].dt.year

# 3. Calculate Company Age (Season Year - Started In)
df['Company Age'] = df['Season_Year'] - df['Started in']

# 4. Handle Edge Cases: Ensure age isn't negative due to same-year starts
# If a company started in 2021 and pitched in 2021, Age is 0 (which is correct)
df['Company Age'] = df['Company Age'].apply(lambda x: max(x, 0))
```

```python

# Forward Mapping: Strings to Numbers
age_mapping = {
    'Young': 0,
    'Middle': 1,
    'Old': 2
}
df['Pitchers Average Age'] = df['Pitchers Average Age'].map(age_mapping)

# Impute missing 'Pitchers Average Age' with the mode of the respective Industry
df['Pitchers Average Age'] = df['Pitchers Average Age'].fillna(
    df.groupby('Industry')['Pitchers Average Age'].transform(
            lambda x: x.mode()[0] if not x.mode().empty else np.nan
        ).round()
)

# # Reverse Mapping: Numbers back to Strings
# reverse_age_mapping = {
#     0.0: 'Young', 
#     1.0: 'Middle', 
#     2.0: 'Old'
# }
# df['Pitchers Average Age'] = df['Pitchers Average Age'].map(reverse_age_mapping)
```

```python

# Calculate the median revenue for each specific industry
industry_medians = df.groupby('Industry')['Yearly Revenue'].transform('median')

# If Revenue is missing AND the company is <= 1 year old -> Fill with 0
# If Revenue is missing AND company is > 1 year old (or age is unknown) -> Fill Industry Median
# If Revenue is NOT missing -> keep the original Revenue


df['Yearly Revenue'] = np.where(  
    df['Yearly Revenue'].isna() & (df["Company Age"] <= 1), 
    0,
    np.where(
        df['Yearly Revenue'].isna(),
        industry_medians, 
        df['Yearly Revenue'] 
    )
)

df['Yearly Revenue'] = df['Yearly Revenue'].fillna(df['Yearly Revenue'].median())
```

```python
# 1. Calculate the median Monthly Sales for each industry as a fallback
industry_monthly_medians = df.groupby('Industry')['Monthly Sales'].transform('median')

# Condition A: If Yearly Revenue is 0, Monthly Sales is 0
# Condition B: If Monthly Sales is missing but Yearly Revenue exists, divide by 12
# Condition C: If it's STILL missing, use the industry median
# Condition D: Keep the original valid Monthly Sales


df['Monthly Sales'] = np.where(

    df['Yearly Revenue'] == 0, 
    0, 
    np.where(  
        df['Monthly Sales'].isna() & (df['Yearly Revenue'] > 0),
        df['Yearly Revenue'] / 12,
        df['Monthly Sales']
    )
)
```

```python
# The "Genius" Save: Convert accidental Markups (>100) into actual Gross Margins
df.loc[df['Gross Margin'] > 100, 'Gross Margin'] = (df['Gross Margin'] / (df['Gross Margin'] + 100)) * 100

# Fill missing values using the specific Industry Median
df['Gross Margin'] = df['Gross Margin'].fillna(
    df.groupby('Industry')['Gross Margin'].transform('median')
)

# The Bulletproof Fallback: Fill any completely empty industries with the overall median
df['Gross Margin'] = df['Gross Margin'].fillna(df['Gross Margin'].median())

df["Gross Margin"] = df["Gross Margin"].astype(float).round(1)
```

```python
# Delete mathematically impossible margins (Net > Gross)
df.loc[df['Gross Margin'] < df['Net Margin'], 'Net Margin'] = np.nan

# Fill the now-empty bad data with the specific Industry Median
df['Net Margin'] = df['Net Margin'].fillna(
    df.groupby('Industry')['Net Margin'].transform('median')
)

# Fallback for completely empty industries
df['Net Margin'] = df['Net Margin'].fillna(df['Net Margin'].median())

# This compares the newly filled Net Margin with the company's actual Gross Margin.
# If the imputed Net Margin accidentally exceeds the Gross Margin, it forces it down.
df['Net Margin'] = np.minimum(df['Net Margin'], df['Gross Margin'])
```

```python
# Fill the blanks with the normal operating standard for that specific industry
df['EBITDA'] = df['EBITDA'].fillna(
    df.groupby('Industry')['EBITDA'].transform('median')
)

# If a whole industry was completely blank, use the global median
df['EBITDA'] = df['EBITDA'].fillna(df['EBITDA'].median())
```

```python
df['Has Patents'] = df['Has Patents'].map({'yes': 1, 'no': 0}).fillna(0).astype(int)
```

```python
df['Bootstrapped'] = df['Bootstrapped'].map({'yes': 1, 'funded': 0}).fillna(0).astype(int)
```

```python
# Calculate both the Median and the Mean for each industry
industry_median = df.groupby('Industry')['SKUs'].transform('median')
industry_mean = df.groupby('Industry')['SKUs'].transform('mean')

# Compare them row-by-row and keep the smaller number
conservative_imputation = np.minimum(industry_median, industry_mean)

# Fill the missing SKU values with that conservative number
df['SKUs'] = df['SKUs'].fillna(conservative_imputation)

# If an entire industry was blank, use the global median
df['SKUs'] = df['SKUs'].fillna(df['SKUs'].median())

# Round any floating-point medians (like 4.5 -> 5.0) 
df['SKUs'] = df['SKUs'].round().astype(int)
```

```python
df["Accepted Offer"].fillna(0, inplace=True)
```

```python
# No Debt was involved: 
# The deal was pure equity (e.g., 50 Lakhs for 5% equity), so an interest rate doesn't exist. This is the case for 90% of the pitches.

# Interest-Free Debt: 
# Occasionally, Sharks will offer debt at 0% interest to help the founders with working capital without burdening them with interest payments.

df["Number of Sharks in Deal"] = df["Number of Sharks in Deal"].fillna(0).astype(int)

df["Total Deal Amount"] = df["Total Deal Amount"].fillna(0)
df["Total Deal Equity"] = df["Total Deal Equity"].fillna(0)
df["Total Deal Debt"] = df["Total Deal Debt"].fillna(0)
df['Debt Interest'] = df['Debt Interest'].fillna(0)

df["Deal Valuation"] = df["Deal Valuation"].fillna(0)
```

```python
display(df.isnull().sum())
display(df.shape)
```

```python
df.drop(['Started in', 'Season Start', 'Season_Year' ,], axis=1, inplace=True)
```
